```python
import pandas as pd

# Define the date range
start_date = "2019-01-01"
end_date = "2020-12-31"

# Generate the list of dates
dates = pd.date_range(start=start_date, end=end_date)

# Create a DataFrame
df = pd.DataFrame({
    'Weekday': [d.strftime('%A') for d in dates], # Sets the day name as the subject
    'Day': [d.strftime('%d/%m/%Y') for d in dates],
    'Week_Number': [d.isocalendar()[1] for d in dates],
    'Month': [d.strftime('%B') for d in dates],
    'Year': [d.year for d in dates]
})

df.to_csv('generated_calendar.csv', index=False)
```
